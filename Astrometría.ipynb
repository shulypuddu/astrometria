```python
# Funci√≥n para generar n√∫meros aleatorios
def generador_numeros_aleatorios(a, c, M, x1, n):
    numeros = []
    x = x1
    for i in range(n):
        x = (a * x + c) % M
        numeros.append(x / M)
    return np.array(numeros)  # Convertir la lista a un array

# Par√°metros del generador de n√∫meros aleatorios
a, c, M, x1 = 1664525, 1013904223, 2**32, 10
N = 1000
K = 10

# Funci√≥n para realizar una caminata aleatoria usando nuestro generador
def random_walk(a, c, M, x1, N, K):
    pasos = []

    for j in range(K):
        dx = []
        dy = []
        x = generador_numeros_aleatorios(a, c, M, x1+j+1, N)
        y = generador_numeros_aleatorios(a, c, M, x1+j, N)
        for i in range(N):
            dx.append(2 * (x[i] - 0.5) * np.sqrt(2))
            dy.append(2 * (y[i] - 0.5) * np.sqrt(2))
        X = np.cumsum(dx)
        Y = np.cumsum(dy)
        pasos.append((X, Y))
    return pasos

# Generar las caminatas
caminata = random_walk(a, c, M, x1, N, K)

plt.figure(figsize=(10, 8))
for i, (X, Y) in enumerate(caminata):
    plt.plot(X, Y)
    plt.scatter(X[0], Y[0])
    plt.scatter(X[-1], Y[-1])
plt.title(f'Caminatas Aleatorias')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

```python
# Calcular el valor de expectaci√≥n de la distancia al origen en funci√≥n de N
def expectacion_distancia_por_paso(caminata, N):
    R = []
    for n in range(1, N+1):
        distancias = [np.sqrt(X[n-1]**2 + Y[n-1]**2) for (X, Y) in caminata]
        R.append(np.mean(distancias))
    return R

# Calcular y graficar la expectaci√≥n de la distancia al origen para cada paso N
R_por_paso = expectacion_distancia_por_paso(caminata, N)

plt.figure(figsize=(10, 6))
plt.plot(range(1, N+1), R_por_paso, label='Expectaci√≥n de $R$')
plt.title('Expectaci√≥n de la Distancia al Origen en funci√≥n de N')
plt.xlabel('N√∫mero de pasos $N$')
plt.ylabel('Expectaci√≥n de la Distancia $R$')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Calcular y graficar la expectaci√≥n de la distancia al origen en funci√≥n de la ra√≠z de N
R_por_paso = expectacion_distancia_por_paso(caminata, N)

plt.figure(figsize=(10, 6))
plt.plot(np.sqrt(np.arange(1, N+1)), R_por_paso, label='Expectaci√≥n de $R$ vs. $\sqrt{N}$')
plt.title('Expectaci√≥n de la Distancia al Origen en funci√≥n de $\sqrt{N}$')
plt.xlabel('$\sqrt{N}$')
plt.ylabel('Expectaci√≥n de la Distancia $R$')
plt.grid(True)
plt.legend()
plt.show()
```

#### Ejercicio 3: Metodo de la Transformada Inversa: Distribucion de Fisher-Tippet

El objetivo en este ejercicio es generar una muestra de N numeros aleatorios tales que sigan la funcion distribucion de probabilidad de Fisher-Tippett, la cual se define como:

###$$ P_{F-T}(x) = \frac{1}{\sigma}t(x)e^{-t(x)},\,\,\,
donde\,\,\,t(x) = \left\{ \begin{array}\\
        [1+ Œæ({\frac{x-\mu}{\sigma}})]^{\frac{-1}{Œæ}} & \mbox{si } \ Œæ\neq0 \\
        e^{\frac{-(x-\mu)}{\sigma}} & \mbox{if } \ Œæ=0 \\
    \end{array}
\right.
$$
  
Con parametros $ \mu=0,\, \xi=0\,\,y\,\, \sigma=1$

Con lo cual, tenemos que

###$$ P_{F-T}(x) = t(x)e^{-t(x)},\,\, t(x)=e^{-x}$$

Por otro lado, se tiene que si U es una variable aleatoria con distribucion uniforme tal que $U\in(0,1)$ y ademas,  P(X) es una distribucion de probabilidad (PDF), con F(X) la funcion acumulada (CDF): $ F(X) = \int_{0}^{X\\}P(s) ds$, entonces, se cumple que la inversa $F^{-1}(U)=X $ genera una variable aleatoria (V.A), con distribucion P(X).
Esto es lo que se conoce como metodo de la transformada inversa, y es una herramienta apropiada para simular muestras con cierta PDF, si se puede calcular de manera analitica la inversa de la CDF.

En este caso, tenemos que:
$$ F(X)=\int_{0}^{X\\}P_{F-T}(s) ds= e^{-t(x)} =exp(-exp(\frac{-x}{\sigma}))$$

$$ F^{-1}(U)= -\sigma\log(-\log(U)) $$

Por lo tanto, utilizando estas funciones y el generador de numeros aleatorios np.random.random(N) podemos realizar una muestra de N valores aleatorios U y mediante el metodo de la Transformada Inversa obtener una muestra de N valores que sigan un la distribucion de Fisher-Tippett

```python
import numpy as np
import matplotlib.pyplot as plt

def fisher_tippett_random(lambda_val, size=1000):
    U = np.random.uniform(0, 1, size)
    return - (1 / lambda_val) * np.log(-np.log(U))

# Parametros
lambda_val = 1.0
size = 10000

# Generamos las nuestras aleatorias
random_samples = fisher_tippett_random(lambda_val, size)

# Calcular la media
sample_mean = np.mean(random_samples)
expected_mean = 0.57721 / lambda_val

print(f"Sample mean: {sample_mean}")
print(f"Theoretical mean: {expected_mean}")

# Gr√°ficos
plt.figure(figsize=(12, 6))

# Histograma
plt.subplot(1, 2, 1)
plt.hist(random_samples, bins=50, color='skyblue')
plt.title('Fisher-Tippett Random Numbers')
plt.xlabel('Valores')
plt.ylabel('Frequencia')

# CDF
sorted_samples = np.sort(random_samples)
cdf = np.arange(1, size+1) / size

plt.subplot(1, 2, 2)
plt.plot(sorted_samples, cdf, color='blue', lw=2)
plt.title('CDF)')
plt.xlabel('Valores')
plt.ylabel('CDF')

plt.tight_layout()
plt.show()
```

#### Ejercicio 4: Metodo de la Transformada Inversa: Distribucion de Poisson

El objetivo de este ejercicio es simular la ocurrencia de N eventos en un proceso de Poisson con una tasa de $\lambda=5/h$, hasta un tiempo T = 3h. Es decir que, si un evento ocurre a un tiempo $t_0$,  la probabilidad de que ocurra otro evento a un tiempo $t_1=t_0+\delta t$, es tal que $$ P_\lambda(\delta t)=\lambda e^{-\lambda \delta t} $$

Para esto nuevamente se puede utilizar el metodo de la Transformada Inversa para obtener una muestra de N $ \delta t $ tal que sigan dicha PDF $ P_\lambda(\delta t) $, donde la CDM $F_\lambda(\delta t)$ y su inversa $F^{-1}_\lambda(U) $ son respectivamente:
$$ F_\lambda(\delta t)=1-e^{-\lambda \delta t} $$

$$ F^{-1}_\lambda(U)=\frac{-1}{\lambda}log(1-U) $$

```python
# Definimos nuestra funci√≥N
def invCPDF(lambda_val, U):
    return -(1 / lambda_val) * np.log(1 - U)

# Parametros
lambda_val = 5  # eventos por hora

N = 17
np.random.seed(5)
U = np.random.random(size=N)
t = invCPDF(lambda_val, U)

t = np.append(0,t)  #le agrego el  tiempo cero
suma_eventos = np.cumsum(t)  #calculo el tiempo acumulado

plt.plot(np.arange(N+1), suma_eventos)
plt.title('Proceso Poisson: Distribuci√≥n de tiempos')
plt.xlabel('Eventos')
plt.ylabel('Tiempo (horas)')
plt.grid(True)
plt.show()
```

#### Ejercicio 7: Experimento de Buffon

El Experimento de Buffon consiste arrojar una aguja de longitud l sobre una mesa que tiene una serie de rayas equiespaciadas una distancia t, y medir la distancia x entre el centro de la aguja y la raya mas cercana a esta, y el aungulo agudo $\theta$ que forma la aguja respecto de la direccion de las rayas. La finalidad del experimento es estimar el valor de $ \pi $.

#### a)
Dado que $ x$ y $ \theta $ son variables independientes, se tiene que la funcion de probabilidad compuesta es:
$$ P(x,\theta)=P_1(x)P_2(\theta) $$

Por otro lado, se tiene que  $ x \in[0,t/2]$, y $ \theta \in [0, \pi/2] $ son todos los posibles valores de $ x $ y $ \theta $, y ademas, todos estos valores son equiprobables  (es decir no hay valores de $ \theta $ o $ x $ privilegiados) con lo cual la PDF de estas variables aleatorias son uniformes: $ P_1(x)=c_1 $ y $ P_2(\theta)=c_2 $.

 A partir de la condicion de normalizacion $ \int_{0}^{t/2\\}c_1 dx = 1 $ tenemos que $ c_1 = \frac{2}{t}$ . Analogamente a partir de $\int_{0}^{\pi/2\\}c_2 dx = 1 $, tenemos que $ c_2= \frac{2}{\pi}$.
Por lo tanto, la funcion densidad de probabilidad de que la aguja caiga con parametros  $ x$ y $\theta $ $P(x,Œ∏)= \frac{4}{t\pi}$

#### b)

Tenemos que $t > l $ (longitud de la aguja es menor a la distancia entre las rayas).
La condicion que se debe cumplir para que la aguja toque una raya es $ x <  \frac{l}{2}\,\sin(\theta)  $.
con los cual la probabilidad de que la aguja toque una raya es:
$$P(x\leq\frac{l}{2}\,\sin(\theta)) = \int_{0}^{\frac{\pi}{2}}\int_{0}^{\frac{l}{2}\,\sin(\theta)} P(x,\theta)\,dxd\theta = \int_{0}^{\frac{\pi}{2}}\int_{0}^{\frac{l}{2}\,\sin(\theta)} \frac{4}{t\pi}\,dxd\theta$$

$$ ‚áí P(x\leq\frac{l}{2}\,\sin(\theta)) = \frac{2l}{t\pi} $$

```python
def simular_buffon(N, l, t):

    cruces = 0  # Contador de cu√°ntas veces la aguja toca una raya

    for _ in range(N):
        x = np.random.uniform(0, t/2)  # Distancia a la raya m√°s cercana
        theta = np.random.uniform(0, np.pi/2)  # √Ångulo agudo con las rayas

        # Condici√≥n para que la aguja cruce una raya
        if (l/2) * np.sin(theta) > x:
            cruces += 1

    # Estimar pi usando la f√≥rmula de Buffon
    pi_est = (2 * l * N) / (t * cruces)
    return pi_est

# Par√°metros
N = 100000  # N√∫mero de simulaciones
l = 29  # Longitud de la aguja
t = 30  # Separaci√≥n entre las rayas (debe ser mayor que l)

# Realizar la simulaci√≥n
pi_estimado = simular_buffon(N, l, t)

# Mostrar resultado
print(f"Estimaci√≥n de pi despu√©s de {N} lanzamientos: {pi_estimado}")

# Comparar con el valor real de pi
print(f"Valor real de pi: {np.pi}")
```

#### Ejercicio 8: Remuestreo Bootstrap

El objetivo de este ejercicio es estimar la varianza de una muestra de N datos y calcular el intervalo de confianza asociado a dicho estimador con un nivel de significancia $ \alpha=0.05$  utilizando el metodo de Bootstrap.

El remuestreo de Bootsrap consiste en elegir aleatoriamente valores de la muestra original permitiendo que los valores se repitan, construyendo asi una nueva muestra del mismo tama√±o que la original, como si se tratara de una nueva medicion de los datos bajo las mismas condiciones. De esta forma se pueden realizar m muestras,(como si fueran m repeticiones del experimento), y asi obtener m estimadores del estadistico (en este caso la varianza). Por ultimo, al realizar un histograma con los estimadores del estadistico se puede llegar a la PDF de estos y obtener un intervalo de confianza con un nivel de significancia $ \alpha$.

```python
# Funciones
def estimar_varianza(datos):
    return np.var(datos, ddof=1)  # ddof=1 para obtener la varianza muestral (n-1)

def bootstrap_varianza(datos, num_resamples=1000, alpha=0.05):

    # Lista para almacenar las varianzas calculadas en cada remuestreo
    varianzas_bootstrap = []

    # Realizar num_resamples remuestreos
    for _ in range(num_resamples):
        # Crear una muestra con reemplazo de los datos
        muestra_bootstrap = np.random.choice(datos, size=len(datos), replace=True)
        # Calcular la varianza de la muestra
        varianza_muestra = estimar_varianza(muestra_bootstrap)
        varianzas_bootstrap.append(varianza_muestra)

    # Calcular percentiles para el intervalo de confianza
    limite_inferior = np.percentile(varianzas_bootstrap, (alpha / 2) * 100)
    limite_superior = np.percentile(varianzas_bootstrap, (1 - alpha / 2) * 100)

    return limite_inferior, limite_superior


# Generar datos
np.random.seed(37)
datos = np.random.normal(loc=0, scale=1, size=100)

# Calcular la varianza original
varianza_original = estimar_varianza(datos)

# Calcular los intervalos de confianza con Bootstrap
lim_inf, lim_sup = bootstrap_varianza(datos, num_resamples=1000, alpha=0.05)

# Graficar el histograma de las varianzas obtenidas con Bootstrap
varianzas_bootstrap = [estimar_varianza(np.random.choice(datos, size=len(datos), replace=True)) for _ in range(1000)]

plt.hist(varianzas_bootstrap, bins=30, alpha=0.7, color='skyblue')
plt.axvline(varianza_original, color='red', linestyle='dashed', linewidth=2, label='Varianza Original')
plt.axvline(lim_inf, color='green', linestyle='dashed', linewidth=2, label=f'IC {100 * (1 - 0.05):.1f}% ')
plt.axvline(lim_sup, color='green', linestyle='dashed', linewidth=2)
plt.xlabel('Varianza')
plt.ylabel('Frecuencia')
plt.legend()
plt.title('Distribuci√≥n de la varianza (Bootstrap) y su intervalo de confianza')
plt.show()
```
