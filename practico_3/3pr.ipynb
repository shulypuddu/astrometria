```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


df0 = pd.read_csv('/mnt/sda2/astrometria/practico_3/P3_shulypuddu.csv')
df= df0[df0['z']>0]
```

```python
# Mostrar paletas una por una
palettes = ["tab10", "tab20", "Set1", "Set2", "Set3", "Paired", "Accent"]
for palette in palettes:
    print(f"Paleta: {palette}")
    sns.palplot(sns.color_palette(palette))
    plt.show()
```

```python
sns.histplot((df['u']-df['r']),bins='auto',alpha=0.75,stat='density')
plt.xlim(0.25,4.5)
plt.xlabel('Índice de color (u-r)')
plt.show()
sns.histplot((df['g']-df['r']),bins='auto',alpha=0.75,stat='density')
plt.xlim(0,1.75)
plt.xlabel('Índice de color (g-r)')
plt.show()
sns.histplot((df['u']-df['g']),bins='auto',alpha=0.75,stat='density')
plt.xlim(0.25,3.75)
plt.xlabel('Índice de color (u-g)')
plt.show()
```
